```python
# Apply the brand mapping to unify brand names in 'combined_data'
combined_data['Brand Name'] = combined_data['Brand Name'].replace(brand_mapping)

# Calculate total Dollar Sales for each brand
brand_sales = combined_data.groupby('Brand Name')['Dollar Sales'].sum().reset_index()

# Identify the top 5 brands by Dollar Sales
top_5_brands = brand_sales.nlargest(5, 'Dollar Sales')['Brand Name']

# Create a new column 'Updated Brand Franchise Name' to categorize brands
combined_data['Updated Brand Franchise Name'] = combined_data['Brand Name'].apply(
    lambda x: x if x in top_5_brands.values else 'Other'
)

# Check the unique values in the new column to ensure proper categorization
print("Unique values in 'Updated Brand Franchise Name':", combined_data['Updated Brand Franchise Name'].unique())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate total sales for each brand and year
brand_year_sales = combined_data.groupby(['Year', 'Updated Brand Franchise Name'])['Dollar Sales'].sum().reset_index()

# Calculate total sales for each year
total_sales_per_year = combined_data.groupby('Year')['Dollar Sales'].sum().reset_index()

# Merge total sales with the brand year sales to compute market share
brand_year_sales = brand_year_sales.merge(total_sales_per_year, on='Year', suffixes=('', '_Total'))
brand_year_sales['Market Share'] = (brand_year_sales['Dollar Sales'] / brand_year_sales['Dollar Sales_Total']) * 100

# Plot the market share trend
plt.figure(figsize=(14, 7))
sns.lineplot(
    data=brand_year_sales,
    x='Year',
    y='Market Share',
    hue='Updated Brand Franchise Name',
    marker='o'
)

# Adding labels and title
plt.title('Market Share Trend Over Time by Brand Franchise', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Market Share (%)', fontsize=12)

# Ensure x-axis only shows integer years
plt.xticks(ticks=brand_year_sales['Year'].unique(), rotation=45)

# Add legend
plt.legend(title='Brand Franchise', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
# Assuming combined_data is your DataFrame
# Categorical Feature Engineering: Classify product volume into ranges
def classify_volume(volume):
    if volume < 8:
        return 'Small (0-7)'
    elif 8 <= volume < 16:
        return 'Medium (8-15)'
    elif 16 <= volume < 32:
        return 'Large (16-31)'
    else:
        return 'Extra Large (32<=)'

# Exclude the 'Other' brand from the data
combined_data = combined_data[combined_data['Updated Brand Franchise Name'] != 'Other']

# Apply the classification function to create the 'Volume Category' column
combined_data['Volume Category'] = combined_data['Total Ounces'].apply(classify_volume)

# Group by 'Brand Name' and 'Volume Category' to calculate the sum of Volume Sales
volume_sales_category = combined_data.groupby(['Updated Brand Franchise Name', 'Volume Category'])['Volume Sales'].sum().reset_index()

# Display the first few rows to check the aggregated data
print(volume_sales_category.head())

# Set up the figure size
plt.figure(figsize=(14, 8))

# Create a bar plot for volume sales by volume category
sns.barplot(
    data=volume_sales_category,
    x='Volume Category',
    y='Volume Sales',
    hue='Updated Brand Franchise Name',
    ci=None  # Remove the confidence interval for better clarity
)

# Add labels and title
plt.title('Volume Sales by Volume Category for Top 5 Brands', fontsize=14)
plt.xlabel('Volume Category', fontsize=12)
plt.ylabel('Total Volume Sales', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Function to calculate VIF
def calculate_vif(data, features):
    X = sm.add_constant(data[features])  # Add constant for intercept
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

# Initial feature set
current_features = X_train_ols.columns.tolist()

# Iteratively remove high-VIF variables
while True:
    vif_df = calculate_vif(X_train_ols, current_features)
    print(vif_df)
    
    # Check the maximum VIF
    max_vif = vif_df["VIF"].max()
    if max_vif < 10:  # Exit if all VIF < 10
        break

    # Find the feature with the highest VIF
    max_vif_feature = vif_df.loc[vif_df["VIF"].idxmax(), "feature"]
    
    # Remove the feature from the current list
    if max_vif_feature in current_features:
        current_features.remove(max_vif_feature)
        print(f"Removing {max_vif_feature} due to high VIF.")
    else:
        print(f"Feature {max_vif_feature} not found in the feature list. Debugging required.")
        break

# Final features after reducing multicollinearity
print("Final features:", current_features)

# Fit the updated regression model
X_train_ols_with_const = sm.add_constant(X_train_ols[current_features])
model = sm.OLS(y_train, X_train_ols_with_const).fit()
print(model.summary())
```

```python
import pandas as pd
import numpy as np
from statsmodels.formula.api import ols
from functools import reduce

# Aggregation Functions (sum for sales/volume, mean for prices, max for distribution)
aggregations = {
    'Base Unit Sales': 'sum',
    'ACV Weighted Distribution': 'max',
    'Dollar Sales No Merch': 'sum',
    'Price per Volume Display Only': 'mean',
    'Incremental Dollars': 'sum',
    'Dollar Sales No Merch': 'sum',
    'Volume Sales': 'sum'
    
}

# List of all brands to analyze
brands = ['GARDEIN', 'MORNINGSTAR FARMS', 'BEYOND MEAT', 'IMPOSSIBLE']

# Aggregation Function for Each Brand
def aggregate_data_for_brand(brand, data):
    brand_data = cleaned_data[cleaned_data['Updated Brand Franchise Name'] == brand]
    return brand_data.groupby('Date').agg(aggregations).rename(
        columns={col: f'{brand} {col}' for col in aggregations.keys()}
    )

# Aggregate the data for each brand
aggregated_dataframes = [aggregate_data_for_brand(brand, cleaned_data) for brand in brands]
```

```python
# Merge all brand dataframes on Date
combined_aggregated_data = reduce(lambda left, right: pd.merge(left, right, on='Date', how='outer'), aggregated_dataframes)

# Apply log transformation to all relevant columns for each brand (adding a small constant to avoid log(0))
for brand in brands:
    combined_aggregated_data[f'log_{brand}_Volume_Sales'] = np.log(combined_aggregated_data[f'{brand} Volume Sales'] + 1)
    combined_aggregated_data[f'log_{brand}_Base_Unit_Sales'] = np.log(combined_aggregated_data[f'{brand} Base Unit Sales'] + 1)
    combined_aggregated_data[f'log_{brand}_ACV_Weighted_Distribution'] = np.log(combined_aggregated_data[f'{brand} ACV Weighted Distribution'] + 1)
    combined_aggregated_data[f'log_{brand}_Price_per_Volume_Display_Only'] = np.log(combined_aggregated_data[f'{brand} Price per Volume Display Only'] + 1)
    combined_aggregated_data[f'log_{brand}_Incremental_Dollars'] = np.log(combined_aggregated_data[f'{brand} Incremental Dollars'] + 1)
    combined_aggregated_data[f'log_{brand}_Dollar Sales No Merch'] = np.log(combined_aggregated_data[f'{brand} Dollar Sales No Merch'] + 1)
```
